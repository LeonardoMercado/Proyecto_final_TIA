```python
import cv2
import numpy as np
```

Ver la cámara a color

```python
cap = cv2.VideoCapture(0)  #Por defecto es cero pero se puede cambiar en funcion del equipo si se tiene multioles camaras.
while(True):
    _, frame = cap.read()
    cv2.imshow('Principal',frame)  #Titulo, imagenes a mostrar
    
    if cv2.waitKey(1) & 0xFF == ord('q'): # Tecla de espera si se oprime la q sale del ciclo.
        break
        
        
cap.release()  #Libera la camara
cv2.destroyAllWindows()  # Cierra todas las ventanas exteriores.
```

Ver la cámara en escala de grises.

```python
cap = cv2.VideoCapture(0)  #Por defecto es cero pero se puede cambiar en funcion del equipo si se tiene multioles camaras.
while(True):
    _, frame = cap.read()  #lee la camara
    
    gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    
    cv2.imshow('Principal',gray)  #Titulo, imagenes a mostrar
    
    
    if cv2.waitKey(1) & 0xFF == ord('q'): # Tecla de espera si se oprime la q sale del ciclo.
        break
        
        
cap.release()  #Libera la camara
cv2.destroyAllWindows()  # Cierra todas las ventanas exteriores.
```

Flag de acceso a la camara 

```python
cap = cv2.VideoCapture(0)  #Por defecto es cero pero se puede cambiar en funcion del equipo si se tiene multioles camaras.
while(cap.isOpened()):   #Si la camara no se abre nunca entra en el ciclo
    _, frame = cap.read()  #lee la camara
    
    gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    
    cv2.imshow('Principal',gray)  #Titulo, imagenes a mostrar
    
    
    if cv2.waitKey(1) & 0xFF == ord('q'): # Tecla de espera si se oprime la q sale del ciclo.
        break
        
        
cap.release()  #Libera la camara
cv2.destroyAllWindows()  # Cierra todas las ventanas exteriores.
```

Método para conocer el tamaño

```python
cap = cv2.VideoCapture(0)  #Por defecto es cero pero se puede cambiar en funcion del equipo si se tiene multioles camaras.
while(cap.isOpened()):
    _, frame = cap.read()  #lee la camara
    
    print(cap.get(cv2.CAP_PROP_FRAME_WIDTH)) #Capruta el ancho del frame
    print(cap.get(cv2.CAP_PROP_FRAME_HEIGHT)) #Captura el alto del frame
    gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    
    cv2.imshow('Principal',gray)  #Titulo, imagenes a mostrar
    
    
    if cv2.waitKey(1) & 0xFF == ord('q'): # Tecla de espera si se oprime la q sale del ciclo.
        break
        
        
cap.release()  #Libera la camara
cv2.destroyAllWindows()  # Cierra todas las ventanas exteriores.
```
